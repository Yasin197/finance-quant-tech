# MLE

Hospital wants to determine risk of not being able to accommodate pateints.

The hospital has 100 beds and 10 years of patient data.
- these are discrete realization over time (that is we can't have a fraction of a person)
  - so it may be reasonable to model this a poisson random variable

$$ p(a) = e^{-\lambda}\frac{\lambda^a}{a!}, a = 0, 1, 2, 3$$

$x_1, x_2, \dots, x_n$ daily patient data

where we model (assume) $X_i \sim Pois(\lambda)$

Note:
- The little x are data points,the big X is denoting the number of patients realized in a day for this particular hospital. The big X has a distribution identically the same as the poisson distribution with some population parameter lambda. So we will try to estimate what should lambda be. So we will be talking about the idea of an `estimator`

***

*Method of Maximum Likelihood Estimation*
- To find lambda, we want to maximise the probability assocaited with the realized random vaibles (X not x) given a particular lambda

$\lambda_{MLE} = argmax_{\lambda} f(X_1, X_2, ..., X_n | \lambda) = argmax_{\lambda}ln(X_1, X_2, 
..., X_n | \lambda))$

$f(X_1, X_2, ..., X_n | \lambda) = \Pi_{i=1}^n f(X_i|\lambda)$
