# Linear Regression in Python 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate some data
x = np.array([np.ones(10), np.arange(0, 10)])
```

```python
x
```

```python
y = x[1, :] + np.random.randn(10)
```

```python
plt.scatter(x[1, :], y);
```

```python
x = x.T
```

```python
x_prod = x.T @ x
```

```python
x_inv = np.linalg.inv(x_prod)
```

```python
y_x = y.T @ x
```

```python
b = x_inv @ y_x
```

```python
# we get 2 values of which the first is the intercept and the other is the slope
b
```

```python
# # plot the line with b
x.shape
```

```python
b.shape
```

```python
# if we compute the matrix product of x,b then we will get the y data
line_data = x @ b
```

```python
plt.scatter(x.T[1, :], y)
plt.plot(x.T[1, :], line_data);
```

# Out of the box solution - sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression().fit(x.T[1, :].reshape(-1, 1), y)
```

```python
b
```

```python
lrb = np.array([lr.intercept_, lr.coef_[0]])
```

```python
lrb
```

```python
# we can round to 5 dp and check if they are the same
np.round(b, 5) == np.round(lrb, 5)
```
